```python
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import numpy as np
```

T-SNE Plot for WGAN

```python
# Load the .npz file
data = np.load("wgan_samples.npz", allow_pickle=True)

# Retrieve the image data and labels
image_data = data["data"]
labels = data["labels"]
```

```python
print(image_data.shape)
image_data = image_data.reshape(-1, 784)
```

```python
tsne = TSNE(n_components=2, random_state=0)
fake_data = tsne.fit_transform(image_data)
```

```python
import pandas as pd
import seaborn as sns
# creating a new data frame which help us in ploting the result data
tsne_data = np.vstack((fake_data.T, labels)).T
tsne_df = pd.DataFrame(data=tsne_data, columns=("Dim_1", "Dim_2", "label"))
sns.FacetGrid(tsne_df, hue="label").map(plt.scatter, 'Dim_1', 'Dim_2').add_legend()
plt.show()
```

T-SNE Plot for DCGAN

```python
# Load the .npz file
data = np.load("dataset_10k.npz", allow_pickle=True)

# Retrieve the image data and labels
image_data = data["images"]
labels = data["labels"]
```

```python
image_data = image_data.reshape(-1, 784)
```

```python
tsne = TSNE(n_components=2, random_state=0)
fake_data = tsne.fit_transform(image_data)
```

```python
# creating a new data frame which help us in ploting the result data
tsne_data = np.vstack((fake_data.T, labels)).T
tsne_df = pd.DataFrame(data=tsne_data, columns=("Dim_1", "Dim_2", "label"))
sns.FacetGrid(tsne_df, hue="label").map(plt.scatter, 'Dim_1', 'Dim_2').add_legend()
plt.show()
```

```python
# Load the MNIST dataset
from tensorflow.keras.datasets import mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Normalize the images
test_images = test_images / 255.0

test_images = test_images.reshape(-1, 784)
```

```python
tsne = TSNE(n_components=2, random_state=0)
real_data = tsne.fit_transform(test_images)
```

```python
# creating a new data frame which help us in ploting the result data
tsne_data = np.vstack((real_data.T, test_labels)).T
tsne_df = pd.DataFrame(data=tsne_data, columns=("Dim_1", "Dim_2", "label"))
sns.FacetGrid(tsne_df, hue="label").map(plt.scatter, 'Dim_1', 'Dim_2').add_legend()
plt.show()
```

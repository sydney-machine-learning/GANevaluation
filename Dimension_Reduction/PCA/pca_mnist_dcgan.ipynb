```python
def pca(fake, real, label):
  print("plot for label: ", label)
  pca1 = PCA(n_components=2)
  pca2 = PCA(n_components=2)

  real_pca = pca1.fit_transform(real)
  fake_pca = pca2.fit_transform(fake)

  # Plot the first PCA scatter plot
  plt.scatter(real_pca[:, 0], real_pca[:, 1], color='blue', label='Real')

  # Plot the second PCA scatter plot
  plt.scatter(fake_pca[:, 0], fake_pca[:, 1], color='red', label='Fake')

  # Add labels and title
  plt.xlabel('Principal Component 1')
  plt.ylabel('Principal Component 2')
  plt.title('PCA Scatter plots of real and fake')

  # Add legend
  plt.legend()

  # Show plot
  plt.show()
```

```python
def make_fake_data(num):
  fake_data = data["images"]
  fake_data_labels = data["labels"]
  fake_data = fake_data.reshape(-1, 28, 28)
  indices = np.where(fake_data_labels==num)[0]
  fake_data = fake_data[indices]
  fake_data = fake_data.reshape(-1, 784)
  return fake_data
```

```python
fake_imgs_zeros = make_fake_data(0)
fake_imgs_ones = make_fake_data(1)
fake_imgs_twos = make_fake_data(2)
fake_imgs_threes = make_fake_data(3)
fake_imgs_fours = make_fake_data(4)
fake_imgs_fives = make_fake_data(5)
fake_imgs_sixes = make_fake_data(6)
fake_imgs_sevens = make_fake_data(7)
fake_imgs_eights = make_fake_data(8)
fake_imgs_nines = make_fake_data(9)
print(images_of_zeros.shape)
print(fake_imgs_zeros.shape)
```

```python
pca(fake_imgs_zeros, images_of_zeros, 0)
pca(fake_imgs_ones, images_of_ones, 1)
pca(fake_imgs_twos, images_of_twos, 2)
pca(fake_imgs_threes, images_of_threes, 3)
pca(fake_imgs_fours, images_of_fours, 4)
pca(fake_imgs_fives, images_of_fives, 5)
pca(fake_imgs_sixes, images_of_sixes, 6)
pca(fake_imgs_sevens, images_of_sevens, 7)
pca(fake_imgs_eights, images_of_eights, 8)
pca(fake_imgs_nines, images_of_nines, 9)
```

```python
# prompt: scatter plot of mnist data  using pca give the labels also legend also
import seaborn as sns
import matplotlib.pyplot as plt
# Load the MNIST dataset
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Normalize the images
test_images = test_images / 255.0

test_images = test_images.reshape(-1, 784)
print(test_images.shape)
# Perform PCA on the training images
pca = PCA(n_components=2)
pca.fit(test_images)

# Transform the test images
test_images_pca = pca.transform(test_images)

plt.figure(figsize=(10, 8))
sns.scatterplot(x=test_images_pca[:, 0], y=test_images_pca[:, 1], hue=test_labels, palette='tab10', legend='full', alpha=0.8)
plt.title('MNIST dataset visualization using PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
```

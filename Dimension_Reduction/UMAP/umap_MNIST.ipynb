```python
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import mnist
from sklearn.preprocessing import StandardScaler
from umap import UMAP
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Flatten images
X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_test_flat = X_test.reshape(X_test.shape[0], -1)

# Scale pixel values to the range [0, 1]
X_train_scaled = X_train_flat.astype('float32') / 255.0
X_test_scaled = X_test_flat.astype('float32') / 255.0
```

```python
umap = UMAP(n_components=2)
X_umap = umap.fit_transform(X_train_scaled)
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(X_umap[:, 0], X_umap[:, 1], c=y_train, cmap='tab10', s=5)
plt.colorbar(label='Digit')
plt.title('UMAP Projection of MNIST Dataset')
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')
plt.show()
```
